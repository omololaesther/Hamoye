```python
# Converting the "Total Charges column" to numeric values
Data["TotalCharges"] = pd.to_numeric(Data["TotalCharges"], errors = "coerce")
```

```python
# Confirming the changed data type
Column_data_types = Data.dtypes
Column_data_types
```

```python
# Handling of missing values
Data['TotalCharges'] = Data['TotalCharges'].fillna(0)
```

```python
# Confirming that there are no missing values
Data.isnull().sum()
```

```python
# Converting the 'Churn' column to binary values, where 'No' is mapped to 0 and 'Yes' is mapped to 1.
Data["Churn"] = Data["Churn"].map({"No" : 0 , "Yes" : 1})
```

```python
Data.head()
```

```python
# Spliting the data into an 80-20 train-test split with a random state of â€œ1â€.
x = Data.drop(columns=["Churn"])
y = Data["Churn"]
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size = 0.2, random_state = 1)
```

```python
# Selecting these features:
categorical = ['gender', 'SeniorCitizen', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'InternetService','OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies','Contract', 'PaperlessBilling', 'PaymentMethod']
categorical_data = Data[categorical]

numerical = ['tenure', 'MonthlyCharges', 'TotalCharges']
numerical_data = Data[numerical]
```

```python
# The numerical features should be scaled using StandardScaler,
# convert the output back to a dataframe and put back the column names.

from sklearn.preprocessing import StandardScaler
column_names =  numerical_data.columns
scaler = StandardScaler()
scaled_features = scaler.fit_transform(numerical_data)

# Convert the scaled features back to a DataFrame
scaled_numerical = pd.DataFrame(scaled_features, columns = column_names)
```

```python
# The categorical features are one-hot encoded using OneHotEncoder(set sparse_output to false),
# convert the output back to a dataframe and put back the column names.

from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse_output = False)
encoded_features = encoder.fit_transform(categorical_data)
column_names = encoder.get_feature_names_out(input_features = categorical_data.columns)

# Convert the one-hot encoded features back to a DataFrame with column names
encoded_categorical = pd.DataFrame(encoded_features, columns = column_names)
```

```python
# Combine scaled numerical and one-hot encoded categorical features into train and test set dataframes
# For train
x_train_combined_data = pd.concat([scaled_numerical, encoded_categorical], axis = 1)
# For test
x_test_combined_data = pd.concat([scaled_numerical, encoded_categorical], axis = 1)
```

```python
# Use scikit learn to train a random forest and extra trees classifier

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.metrics import accuracy_score

x = x_test_combined_data.drop(columns=['PaymentMethod_Credit card (automatic)'])
y = x_test_combined_data["PaymentMethod_Credit card (automatic)"]

x_train,x_test,y_train,y_test = train_test_split(x, y, test_size = 0.2, random_state = 1)

RF_Classifier = RandomForestClassifier(n_estimators = 100, random_state = 1)
RF_Classifier.fit(x_train, y_train)
RF_Pred = RF_Classifier.predict(x_test)
RF_Accuracy = accuracy_score(y_test, RF_Pred)
print('RF_Accuracy: ', RF_Accuracy)

ET_Classifier = ExtraTreesClassifier(n_estimators = 100, random_state = 1)
ET_Classifier.fit(x_train, y_train)
ET_Pred = ET_Classifier.predict(x_test)
ET_Accuracy = accuracy_score(y_test, ET_Pred)
print('ET_Accuracy: ', ET_Accuracy)
```

```python
# use xgboost and lightgbm to train an extreme boosting model and a light gradient boosting model

import xgboost as xgb
import lightgbm as lgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

x = x_test_combined_data.drop(columns=['PaymentMethod_Credit card (automatic)'])
y = x_test_combined_data["PaymentMethod_Credit card (automatic)"]

x_train,x_test,y_train,y_test = train_test_split(x, y, test_size = 0.2, random_state = 1)

XGB_Classifier = xgb.XGBClassifier(n_estimators = 100, random_state = 1)
XGB_Classifier.fit(x_train, y_train)
XGB_Pred = XGB_Classifier.predict(x_test)
XGB_Accuracy = accuracy_score(y_test, XGB_Pred)
print('XGB_Accuracy :', XGB_Accuracy)

LGB_Classifier = lgb.LGBMClassifier(n_estimators = 100, random_state = 1)
LGB_Classifier.fit(x_train,y_train)
LGB_pred = LGB_Classifier.predict(x_test)
LGB_Accuracy = accuracy_score(y_test, LGB_pred)
print('LGB_Accuracy :', LGB_Accuracy)
```
