```python
# Question 8
FBS3 = FBS[["Element Code", "Y2014", "Y2015", "Y2016", "Y2017", "Y2018"]]
# Calculate the correlation coefficients between 'Element Code' and the years
correlation_with_elementcode = FBS3.corr()['Element Code']

# Find the variable with the least absolute correlation coefficient
least_correlated_year = correlation_with_elementcode.abs().idxmin()

print("Year with the least correlation with 'Element Code':", least_correlated_year)
print("Correlation coefficient:", correlation_with_elementcode[least_correlated_year])
```
